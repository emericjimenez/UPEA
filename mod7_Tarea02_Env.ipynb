```python
import numpy as np
import pandas as pd
import seaborn as sns
import csv
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
#Identificamos las columnas que tendra y leemos kardex,csv
colnames=['ci_estudiante', 'ci_est_dpto', 'materia', 'ci_docente', 'ci_doc_dpto','periodo', 'nota', 'observaciones'] 

df0 = pd.read_csv('kardex.csv', sep='\s', names=colnames, engine='python')

# Eliminamos comillas y otros simbolos
df0['ci_estudiante'] = df0['ci_estudiante'].str.replace('"', '', regex=False)
df0['observaciones'] = df0['observaciones'].str.replace('"', '', regex=False)
df0['nota'] = df0['nota'].str.replace('"', '', regex=False)
df0['ci_est_dpto'] = df0['ci_est_dpto'].str.replace('.', '', regex=False)
df0['ci_doc_dpto'] = df0['ci_doc_dpto'].str.replace('.', '', regex=False)

# Guardamos en un nuevo archivo
df0.to_csv("k.csv",index=False)
```

```python
#Leemos el nuevo archivo
df = pd.read_csv('k.csv')
df.head()
```

```python
#Estadisticas
df.describe()
```

```python
#Hay que arreglar las columnas = materia, periodo, nota y observaciones
df.isnull().sum()
```

```python
#Empezamos con observaciones y vemos que la columna nota, periodo, ci_doc_dpto, ci_docente estan mal ubicadas
filas_con_nulos = df[df['observaciones'].isna()]
filas_con_nulos
```

```python
#Entonces copiamos los valores de una columna a otra para corregir
result_nota = df.loc[df['nota'].isin(['Aprobado'])]                
df.loc[result_nota.index, 'observaciones'] = result_nota.loc[result_nota.index, 'nota']
df.loc[result_nota.index, 'nota'] = result_nota.loc[result_nota.index, 'periodo']
df.loc[result_nota.index, 'periodo'] = result_nota.loc[result_nota.index, 'ci_doc_dpto']
df.loc[result_nota.index, 'ci_doc_dpto'] = 'LP'
df.loc[result_nota.index, 'ci_docente'] = 0

result_nota = df.loc[df['nota'].isin(['Reprobado'])]                
df.loc[result_nota.index, 'observaciones'] = result_nota.loc[result_nota.index, 'nota']
df.loc[result_nota.index, 'nota'] = result_nota.loc[result_nota.index, 'periodo']
df.loc[result_nota.index, 'periodo'] = result_nota.loc[result_nota.index, 'ci_doc_dpto']
df.loc[result_nota.index, 'ci_doc_dpto'] = 'LP'
df.loc[result_nota.index, 'ci_docente'] = 0

#Las columnas siguientes las convertimos en float e int para analizar mejor
df['nota'] = df['nota'].astype('float64')
df['ci_docente'] = df['ci_docente'].astype('int64')

#df['observaciones'].unique()

df.describe()

```

```python
#Verificamos
#filas_con_nulos = df[df['observaciones'].isna()]
filas_con_nulos = df[df['nota'].isna()]
filas_con_nulos
```

```python
#Corregimos la columna nota, usamos la media de los aprobados y la media de los reprobados para completar los valores
#faltantes en base a si en la columna observaciones esa fila esta aprobada o reprobada
filas_con_nulos = df[df['nota'].isna()]
aprobados = df[df['nota'] > 50]
reprobados = df[df['nota'] < 51]
aprobados['nota'].mean()
reprobados['nota'].mean()

filas_con_nulos.loc[filas_con_nulos['observaciones'] == "Aprobado", 'nota'] = aprobados['nota'].mean()
filas_con_nulos.loc[filas_con_nulos['observaciones'] == "Reprobado", 'nota'] = reprobados['nota'].mean()

df.loc[filas_con_nulos.index, 'nota'] = filas_con_nulos.loc[filas_con_nulos.index, 'nota']

filas_con_nulos = df[df['nota'].isna()]
filas_con_nulos
df.isnull().sum()
```

```python
#Vemos la columna materia y sus valores nulos
#filas_con_nulos = df[df['periodo'].isna()]
filas_con_nulos = df[df['materia'].isna()]
filas_con_nulos
```

```python
#luego de analizar los valores faltantes respecto a docente y alumno llegamos a llenar los valores por los sgtes:
def procesar_nan(peri, ci_doc, ci_est, valor):
    result_mat = df.loc[df['periodo'].isin([peri])]
    alumno = result_mat[(result_mat['ci_estudiante']==ci_est)]
    #print(alumno)
    prof = result_mat[(result_mat['ci_docente']==ci_doc)]
    #print(prof)
    df.loc[alumno.index, 'materia'] = valor

procesar_nan('I-2000', 8321673, 6800465, 'INF325')
procesar_nan('I-2000', 8321673, 6838626, 'INF325')
procesar_nan('II-2000', 4312014, 8301162, 'MAT270')
procesar_nan('II-2000', 6056909, 6869689, 'INF120')
procesar_nan('II-2005', 6852438, 4965844, 'EST280')   
procesar_nan('II-2007', 6182990, 6869126, 'EST140')
procesar_nan('II-2011', 6761652, 8305840, 'MAT320')
procesar_nan('I-2012', 4758753, 6900933, 'INF324')
```
