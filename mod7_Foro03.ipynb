```python
import pandas as pd
```

```python
df = pd.read_csv("estudiantes03.csv", delimiter=';', index_col=0)
```

```python
#Verificamos cuantos nulos hay en Depto
filas_con_nulos = df[df['Depto'].isna()]
filas_con_nulos
#df.isnull().sum()
```

```python
#Solo hay 2 filas con columna depto = nan, así que las eliminamos y luego verificamos
df = df.dropna(subset=['Depto'])
df.isnull().sum()
```

```python
#Vemos si existe algun problema con los nombres de los departamentos
df["Depto"].unique()
#df
```

```python
#Como existe palabras mal escritas las corregimos y verificamos
#Como El Alto pertenece al departamento de La Paz lo modificamos
df.loc[df["Depto"] == "EL ALTO", "Depto"] = "LA PAZ"
df.loc[df["Depto"] == "COCABAMBA", "Depto"] = "COCHABAMBA"
df.loc[df["Depto"] == "SANTA CRUS", "Depto"] = "SANTA CRUZ"
df["Depto"].unique()
```

```python
#Escogemos 3 departamentos para crear las nuevas columnas
#Departamentos = ['LA PAZ', 'POTOSI', 'COCHABAMBA', 'TARIJA', 'SANTA CRUZ', 'ORURO', 'PANDO', 'CHUQUISACA', 'BENI']
Departamentos = ['LA PAZ', 'COCHABAMBA', 'SANTA CRUZ']
```

```python
#Usamos numpy para hacer la busquqeda y hacemos el llenado de filas
import numpy as np
df1 = df.drop(columns=["Depto"])
for departamento in Departamentos:
    df1[departamento] = np.where(df['Depto'] == departamento, 1, 0)

#Vemos los resultados
df1
```
