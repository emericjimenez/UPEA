```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import re
nltk.download('punkt')
new_sw = ["dio", "en", "ser", "Los", "ver", "la"]
stop_words = nltk.corpus.stopwords.words('Spanish')
stop_words.extend(new_sw)
```

```python
file = open('art.txt', 'r')
read_file = file.read()
read_file =re.sub(r'[.,"\'-?:!;]', '', read_file)
text = nltk.Text(nltk.word_tokenize(read_file))
new_words = []
for word in text:
    if word not in stop_words:
        new_words.append(word)
frecWord = []
for w in new_words:
    frecWord.append(new_words.count(w))
```

```python
import pandas as pd
df = pd.DataFrame()
df['Repeticiones'] = frecWord
df['Palabra'] = new_words
df = df.sort_values('Repeticiones',ascending=False)
df = df.drop_duplicates()
df.head(40)
```

```python
import numpy as np
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
plt.rc('xtick', labelsize=20) 
plt.rc('ytick', labelsize=20) 
fig.set_figheight(7)
fig.set_figwidth(18)
ax.bar(df['Palabra'].head(18),df['Repeticiones'].head(18))
ax.set_ylabel('Palabras usadas')
ax.set_title('Análisis')
plt.xticks(rotation=45)
plt.show()
```
